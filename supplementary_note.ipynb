## 補足：plt.figure() の引数 figsize, dpi と、描画領域のサイズ figure size について

```python
# デフォルトの figsize
print(plt.rcParams["figure.figsize"])
```

```python
# デフォルトの dpi
print(plt.rcParams["figure.dpi"])
```

```python
# デフォルトのfigure size
fig = plt.figure()
```

```python
# figsize、dpi ともにデフォルトのまま、画像を表示をする。
plt.figure(figsize=(6, 4), dpi=72, edgecolor="coral", linewidth=2)
plt.imshow(x_train[0, :, :, :])
plt.show()
```

**例１. dpi の値を大きくする。**

```python
# dpiの値 を 72 → 108 にする。figure size は、648×432 となる。
plt.figure(figsize=(6, 4), dpi=108, edgecolor="coral", linewidth=2)
plt.imshow(x_train[0, :, :, :])
plt.show()

# 解像度を高くしながら、縦横に1.5倍拡大していると言える。
```

**例２. figsize の値を大きくする。**

```python
# figsizeの値 を (6, 4) → (9, 6) にする。figure size は、648×432 となる。
plt.figure(figsize=(9, 6), dpi=72, edgecolor="coral", linewidth=2)
plt.imshow(x_train[0, :, :, :])
plt.show()
# 解像度は同じまま、縦横に1.5倍、拡大していると言える。
```

**例３. figure size はデフォルトのまま、figsize は小さく、dpi は大きくする。**

```python
# figure size はデフォルトの 432×288 のまま、
# figsize を(6, 4) → (3.6, 2.4) に、dpiを 72 → 120 にする。
plt.figure(figsize=(3.6, 2.4), dpi=120, edgecolor="coral", linewidth=2)
plt.imshow(x_train[0, :, :, :])
plt.show()
```

**例４. figure size はデフォルトのまま、figsize は大きく、dpi は小さくする。**

```python
# figure size はデフォルトの 432×288 のまま、
# figsize を(6, 4) → (9, 6) に、dpiを 72 → 48 にする。
plt.figure(figsize=(9, 6), dpi=48, edgecolor="coral", linewidth=2)
plt.imshow(x_train[0, :, :, :])
plt.show()
```
