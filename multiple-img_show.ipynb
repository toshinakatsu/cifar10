## ラベルに対応する画像を表示する

### あるラベルに対応する画像を表示する

**例１. ラベル9の画像の中から、2枚目の画像を表示する。**

---

```python
# ラベル9の画像のインデックスを取得する。
np.where(y_train == 9)[0]
```

```python
# ラベル9の画像の、2個目のインデックスを取得する。
np.where(y_train == 9)[0][1]
```

```python
# ラベル9の画像の中から、2枚目の画像を表示する。
plt.imshow(x_train[np.where(y_train == 9)[0][1], :, :, :])
```

**一般的に、ラベル i の n 枚目の画像を表示するには、次のように書ける。**  

\begin{equation}
  plt.imshow\:(x\verb|_|train\:[\;np.where\:(y\verb|_|train == i)[\:0\:][\:n - 1\:],\;\;:,\;\;:,\;\;:\;])
\end{equation}

**例. ラベル 6 の 1 枚目の画像を表示する**

---

```python
plt.imshow(x_train[np.where(y_train == 6)[0][0], :, :, :])
```

**ラベル 6 の 1 枚目の画像が表示できていることが分かります。**

**例２. ラベル 9 の画像の中から、5 枚目までの画像を表示する。**

---

```python
# 初期設定 -- ここから
category = ['飛行機', '自動車', '鳥', '猫', '鹿', '犬', 'カエル', '馬', '船', 'トラック']
ncols=5
# 初期設定 -- ここまで
 
# squeeze=Falseとしておけば、ncols=1でもaxes[i,j]と書ける。
fig, axes = plt.subplots(nrows=1, ncols=ncols, figsize=(5, 8), dpi=100, squeeze=False)

# ラベル9の画像のインデックスを取得する。
label_9_index = np.where(y_train == 9)[0]

for j in range(ncols):
    
    # 1行目、j+1列目に、
    # ラベル9の画像の中から、j+1番目の画像を表示する。 
    axes[0,j].imshow(x_train[label_9_index[j], :, :, :])

    # 目盛りを非表示にする。
    axes[0, j].set_xticks([])
    axes[0, j].set_yticks([])

    # 1列目の画像のみy軸ラベルをつける。
    if j == 0:
        axes[0,j].set_ylabel(f'ラベル{9}:{category[9]}', labelpad = 40, rotation="horizontal")

plt.show()
```

### 各ラベルに対応する画像を複数表示する

**例. 各ラベルに対応する画像を 5 枚目まで表示する。**  
・y_train から、ラベルのインデックスを取得し、x_trainの画像指定に利用する。 → arg何とかを使う。  
・y軸ラベルは、1列目の画像を表示するときだけつければよい。→ if文を使う。     
・関数化しておく。→ 予測した画像との比較に使えそう。k-nnなど。

```python
# 初期設定 -- ここから
category = ['飛行機', '自動車', '鳥', '猫', '鹿', '犬', 'カエル', '馬', '船', 'トラック']
nrows=10
ncols=5
# 初期設定 -- ここまで
 
# squeeze=Falseとしておけば、ncols=1でもaxes[i,j]と書ける。
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(3.75, 6.25), dpi=96, squeeze=False)

for i in range(nrows):

    # ラベルiの画像のインデックスを取得する。
    label_i_index = np.where(y_train == i)[0]

    for j in range(ncols):
    
        # i+1行目、j+1列目に、
        # ラベルiの画像の中から、j+1番目の画像を表示する。 
        axes[i,j].imshow(x_train[label_i_index[j], :, :, :])

        # 目盛りを非表示にする。
        axes[i, j].set_xticks([])
        axes[i, j].set_yticks([])

        # 1列目の画像のみy軸ラベルをつける。
        if j == 0:
            axes[i,j].set_ylabel(f'ラベル{i}:{category[i]}', labelpad = 40, rotation="horizontal")

plt.show()
```

**ラベル i の画像の中から、6 枚目～10 枚目までの画像を表示させるには、**  

\begin{equation}
        label\verb|_|i\verb|_|index\:[\;j\;]
\end{equation}

**の j を j + 5 に置き換えることで、ラベル i の 6 番目のインデックスから取得すればよい。**





**また、ランダムに画像を表示させるには、j を np.random.randint(5000) に置き換えることで、**
